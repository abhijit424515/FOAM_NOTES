# Data-types and Variables


Python has the following basic data-types
- `Numbers` - Integers and Floats
- `Booleans` - `True` or `False`
- `Strings`

`type()` function returns the type of a variable.

```python
# Examples

var_31 = 31
var_float = 3.23
name = "John Doe"
z = False

print(type(var_31))
print(type(var_float))
print(type(name))
print(type(z))
```

### Dynamic Typing
Types are not fixed in Python. They are inferred during runtime and can be changed for the same variable.

```python
a = 3
print(type(a))

a = "Now I am a string"
print(type(a))
```

### Operations


```python
# Numbers

var = 3.5

# Operators in Python
print("Sum", var + 1)
print("Product", var * 2)
print("Diff", var - 2)
print("Division", var / 2)
print("Integer division", var // 2)
print("Remainder", var % 2) 
print("Exponentiation", var ** 3)
```

```python
var *= 3  # var = var*3
print(var)
```

```python
var += 4  # var = var+4
print(var)
```

```python
# Booleans

a = True
b = False

print(type(a))

print("AND", a and b)
print("OR", a or b)
print("NOT", not a)
print("NOT", not 0) # only 0 is considered False by default
print("NOT", not -1) # -1 is considered True
```

You can obtain the length of a list using `len` function.

Appending and popping elements out

```python
diverse_arr = ["Alpha", 3, True, [3, 4, 5]]
diverse_arr.append(4) # add '4' at the last index (new index)
print(diverse_arr)
print(type(diverse_arr[4]))

diverse_arr.pop() # remove the last value
print(diverse_arr)
diverse_arr.pop(2) # remove the value at index 2
print(diverse_arr)

print(diverse_arr.pop(1)) # print the popped value at index 1
print(diverse_arr) # last line had actually modified the list permanently instead of temporarily
```

You often need a list starting from some number to another, with each element exceeding its previous one by a certain difference, basically an AP. Use `range` function.

```python
# Range starting from 0
print(range(10))

print(type(range(10)))

x = list(range(10))
print(x)
```

```python
start = 4
end = 26 # 26-2 = 24 will be the value at the last index
diff = 2
# You can skip the value of diff. In that case, it is defaulted to 1. 
x = range(start, end, diff)
print(x)
print(list(x))
```

```python
start = 3
end = 28 # end value works as a boundary, while start value is always included in the list
diff = 2
# You can skip the value of diff. In that case, it is defaulted to 1. 
x = range(start, end, diff)
print(x)
print(list(x))
```

### Array Slicing
You can select a section of a list using the start (inclusive) and end (exclusive) indices.

```python
arr = list(range(100))
start = 3
end = 28
diff = 2
print(arr)
print(arr[start:end:diff])
```

```python
b = [3, 5, 6, 7, 8, 2, 1, 0, 3, 2]

print(b[1:8:2])
print(b[1:6:2])
print(b[5:])
print(b[:3])
print(b[6:1]) # rather than put up an error, it prints an empty list
```

**Question** - Can you use this idea of splicing to reverse a list?

```python
print(b)
print(b[::1]) # index order : 0,1,2,...
print(b[::-1]) # index order : -1,-2,-3,... 
print(b[0:3:-1]) # index order : none
print(b[3:0:-1]) # index order : 3,2,1
```

# Loops
Loops allow you to do anything iteratively.

```python
lis = [3, 4, 5, "animal", True]
for value in lis:
    print(2*value)
```

### Continue, Break
1. What if you want to skip some iterations?
2. What if you want to stop iteration once some condition is met?

```python
lis = [3, 4, 5, "animal", True]
for value in lis:
    print("Checked", value)
    if value == 4:
        continue
    print("Passed", value)
```

```python
for value in lis:
    if value == 5:
        break
    print(value)

print("XYZ")
```

```python
def multiply(a, b):
    return a*b

def add(a, b):
    print(a+b)
    # return a+b            # forgot to return anything, so prints None

x = multiply(3, 4)
print(x)

y = add(3, 4)
print(y)

z = None
print(z)
z=3
print(z)
```

You might wish to write functions with default arguments such that you do not need to provide these arguments everytime you want that default value to be used!

```python
def add(a, b = 4):
    print(a)
    print(b)
    print( a+b )
    print("Function call over")

a = 9
add (3, 5)
add (2)
add (a=5, b=3)  # temporary changes to variable a
add (b=3, a=5)  # temporary changes to variable a         # order didn't matter here, as we were targetting the variables
add (a=2)
# add (b=6)     # this would show up an error, as no value was given for a

print(a)        # temporary changes to variable a
```

**Exercise** - Try finding out the approximate range for the `random.random()` function.

```python
min_value = None
max_value = None

for i in range(1000000):
    z = r.random()
    if min_value is None:
        min_value = z
    else:
        min_value = min(min_value, z)
    if max_value is None:
        max_value = z
    else:
        max_value = max(max_value, z)
print(min_value, max_value)
```
